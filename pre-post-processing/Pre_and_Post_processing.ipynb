```python
import re # only to test

# data structure to pass
# [
#     text,
#     [
#     regex,
#     replacement_txt
#     ]
# ]
# flags=re.IGNORECASE
def replace_text(data_struct):
    text = data_struct[0]
    regex = data_struct[1][0]
    replacement_txt = data_struct[1][1]
    # print(text, regex, replacement_txt)
    # return re.sub(regex, replacement_txt, text, flags)

    # Find all matches of the regex in the text
    matches = re.findall(regex, text, flags=re.I | re.M)

    # Print the matches
    if matches:
        print("Regex matches:", matches)
    # else:
    #     print("No regex matches found.")

    return re.sub(regex, replacement_txt, text, flags=re.I | re.M)

def run_regex(txt, strs):
    for regex_struct in strs:
      # print(regex_struct)
      data_struct = [
          txt,
          regex_struct
      ]
      # print(data_struct)
      txt = replace_text(data_struct)
    return txt

def run_regex_post(txt, strs=post_strs):
    return run_regex(txt, strs)

def run_regex_pre(txt, strs=pre_strs):
    return run_regex(txt, strs)
```

```python
def process_csv(input_file, output_file=None, fun=run_regex_pre):
    """
    Process the source_text column of a CSV file using regex transformations

    Args:
        input_file (str): Path to input CSV file
        output_file (str, optional): Path to save processed CSV. If None, adds '_processed' suffix
    """
    # Read the CSV file
    df = pd.read_csv(input_file)

    if 'source_text' not in df.columns:
        raise ValueError("CSV file must contain a 'source_text' column")

    # Apply pre-processing
    df['processed_text'] = df['source_text'].astype(str).apply(fun)

    # Apply post-processing
    df['processed_text'] = df['processed_text'].apply(fun)

    # Save processed file
    if output_file is None:
        output_file = input_file.rsplit('.', 1)[0] + '_preprocessed.csv'

    df.to_csv(output_file, index=False)
    print(f"Processed file saved to: {output_file}")

    # Return sample of processed texts
    return df[['source_text', 'processed_text']].head()

file_path = 'train.csv'
# /teamspace/studios/this_studio/
try:
    sample_results = process_csv(file_path)
    print("\nSample of processed texts:")
    print(sample_results)
except Exception as e:
    print(f"Error processing file: {str(e)}")

```

```python
def compare_source_text(file1, file2):
  """
  Compares the 'source_text' column between two CSV files.

  Args:
    file1: Path to the first CSV file.
    file2: Path to the second CSV file.
  """
  df1 = pd.read_csv(file1)
  df2 = pd.read_csv(file2)

  if 'source_text' not in df1.columns or 'source_text' not in df2.columns:
    raise ValueError("Both CSV files must contain a 'source_text' column.")

  # Compare the values in the 'source_text' column
  diff = df1['source_text'] != df2['source_text']

  # Print rows with differences
  if diff.any():
    print("Differences found in 'source_text' column:")
    for index, row in df1[diff].iterrows():
      print(f"Row {index}:")
      print(f"  {file1}: {row['source_text']}")
      print(f"  {file2}: {df2.loc[index]['source_text']}")
  else:
    print("No differences found in 'source_text' column.")

# Example usage:
compare_source_text('train.csv', 'train_preprocessed.csv')
```
